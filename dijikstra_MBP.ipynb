# Graph Genertaion 

```python
import random
class Node(object):
    def __init__(self,v,w):
        self.vertex=v
        self.weight=w
```

```python
def graph_generation():
    '''
    number of vertices is 5000 numbered from 1 to 4999
    edge weights vary from 1 to 1000
    adjacency list for every vertex is list of 'Node' objects (contains vertex name and weight of the edge)
    '''
    
    degree = [int(random.normalvariate(6,1)) for i in range(5000)]
    
    #can be deleted
    #to check the average of the degree of edges 
    sum=0
    for i in range(5000):
        sum+=degree[i]
    print("Average degree of the edges is : ", sum/5000)
    print("\n")
    
    #creating a cycle, connecting the whole graph
    adj_list = [[] for _ in range(5000)]
    for i in range(1,5000):
        edge_weight = random.randint(1,1000)
        adj_list[i].append(Node(i-1,edge_weight))
        degree[i]-=1
        adj_list[i-1].append(Node(i,edge_weight))
        degree[i-1]-=1
        
    edge_weight = random.randint(1,1000)    
    adj_list[0].append(Node(4999,edge_weight))
    degree[0]-=1
    adj_list[4999].append(Node(0,edge_weight))
    degree[4999]-=1
    
    print("Graph is connected with a cycle")
    print("\n")
    
    #creating edges randomly until degree[vertex] becomes 0
    for vertex in range(5000):
        ##print("Currently checking vertex ", vertex)
        existence={}
        num_try=0
        for neigh in adj_list[vertex]:
            existence[neigh.vertex]=1
            
        while degree[vertex]>0:
            ##print('num_try: ',num_try)
            new_neighbor = random.randint(0,4999)
            if degree[new_neighbor]>0 and new_neighbor not in existence:
                edge_weight=random.randint(1,1000)
                adj_list[vertex].append(Node(new_neighbor,edge_weight))
                degree[vertex]-=1
                adj_list[new_neighbor].append(Node(vertex,edge_weight))
                degree[new_neighbor]-=1

                
            elif num_try>=5 and new_neighbor not in existence:
                edge_weight=random.randint(1,1000)
                adj_list[vertex].append(Node(new_neighbor,edge_weight))
                degree[vertex]-=1
                adj_list[new_neighbor].append(Node(vertex,edge_weight))
                degree[new_neighbor]-=1
                
                num_try=0
                
                
            else:
                num_try+=1    
                
            existence[new_neighbor]=1 
    print("Graph Creation is completed")
    print("\n")
    return adj_list
```

# Dijikstra's for MBP Without Heap

```python
class dijikstra_noHeap(object):
    def __init__(self,x):
        self.adj_list = x
        

    def print_path(self,s,t,dad):
        
        stack=[]
        while t and t!=s:
            #print(t)
            stack.append(t)
            #print('type:',type(t))
            t=dad[t]
            
        path=str(s)+"-->"
        while len(stack)>1:
            path=path+str(stack.pop())+"-->"
        if len(stack)==1:
            path+=str(stack.pop())
        return path    


    def dijikstra_no_heap(self,s,t): 
        status=['unseen']*5000
        dad=['']*5000
        fringe_weight=[-float('Inf')]*5000
        weight=[-float('INF')]*5000
        
        fringe=[]
        status[s]='intree'
        
        for neigh in self.adj_list[s]:
            status[neigh.vertex]='fringe'
            fringe.append(neigh.vertex)
            weight[neigh.vertex]=neigh.weight
            fringe_weight[neigh.vertex]=weight[neigh.vertex]
            dad[neigh.vertex]=s

        while fringe:
            best_fringe = fringe_weight.index(max(fringe_weight))
            #remove it from fringe array
            fringe.remove(best_fringe)
            #add the precessed vertex into the tree
            status[best_fringe]='intree'
            #remove processed vertex weight from the array    
            fringe_weight[best_fringe]=float('-INF')

            for neigh in self.adj_list[best_fringe]:
                if status[neigh.vertex] == 'unseen':
                    status[neigh.vertex]='fringe'
                    fringe.append(neigh.vertex)
                    weight[neigh.vertex]=min(weight[best_fringe],neigh.weight)
                    fringe_weight[neigh.vertex]=weight[neigh.vertex]
                    dad[neigh.vertex]=best_fringe
                    
                elif status[neigh.vertex]=='fringe' and weight[neigh.vertex]<min(weight[best_fringe],neigh.weight):
                    weight[neigh.vertex]=min(weight[best_fringe],neigh.weight)
                    fringe_weight[neigh.vertex]=weight[neigh.vertex]
                    dad[neigh.vertex]=best_fringe

        path=self.print_path(s,t,dad)        

        return path,weight[t]        
```

# Heap Implementation

```python
#implmentation of max heap functions : heappush, heappop, heapmax
#a list to maintain the heap structure using weights of the edges
#a list to maintain the node indices of the nodes in the heap
#a list to mantain the heap indices of the nodes in the heap
#size of the heap before the operation has performed
#item to be inserted
#node to be inserted
#index of the heap to be deleted

class heap(object):
    def __init__(self):
        self.D=[-float('INF')]*5000
        self.H=[]
        self.size=0
        self.node_heap=[None]*5000
    
    def insert_heap(self, D_dij, item, node):
        ''' 
        H : heap, list that maintains mapping between heap index to node index
        D : values of each vertex
        node_heap : list that maintains mapping between node index to heap index
        item : weight of the node to be inserted
        node : node to be inserted
        size : size of the heap before insertion
        '''
        self.D=D_dij
        
        self.H.append(node)
        self.D[node]=item
        self.node_heap[node]=self.size
        
        self.size+=1
        
        last_index=self.size-1
        parent=(last_index-1)//2
        child=last_index
        
        print("Heap Insertion begins")

        while parent>=0 and self.D[self.H[parent]]<self.D[self.H[child]]:
            print("Insertion heap property violated")
            self.node_heap[self.H[parent]], self.node_heap[self.H[child]] = self.node_heap[self.H[child]], self.node_heap[self.H[parent]]
            self.H[parent], self.H[child] = self.H[child], self.H[parent]
            child=parent
            parent=(parent-1)//2
        print("Heap Insertion completed--------------------------------")    
        return self.H, self.D





    def delete_heap(self, D_dij, v):
        ''' 
        H : heap, list that maintains mapping between heap index to node index
        D : values of each vertex
        node_heap : list that maintains mapping between node index to heap index
        ind : index to be deleted
        size : size of the heap before insertion
        '''
        self.D=D_dij
        
        ind=self.node_heap[v]
        
#         if len(H)==1:
#             node_heap[H[0]]=None
#             H.remove(H[0])
#             D[H[0]]=float('-INF')
#             return H, D, node_heap

        last_index=self.size-1
    
        self.node_heap[self.H[last_index]]=ind
        print("INDEX TO BE DELETED IS ", ind)
        self.H[ind]=self.H[last_index]

        self.D[v]=float('-INF')
        self.node_heap[v]=None
        
        self.size-=1

        parent=(ind-1)//2
        child_1, child_2 =(2*ind)+1, (2*ind)+2
        

        
        #heapify up
        if parent>=0 and self.D[self.H[ind]]>self.D[self.H[parent]]:
            print("HEAPIFY UP BEGINS")
            child=ind
            while parent>=0 and self.D[self.H[parent]]<self.D[self.H[child]]:
                #D[H[parent]], D[H[child]] = D[H[child]], D[H[parent]]
                self.node_heap[self.H[parent]], self.node_heap[self.H[child]] = self.node_heap[self.H[child]], self.node_heap[self.H[parent]]
                self.H[parent], self.H[child] = self.H[child], self.H[parent]
                child=parent
                parent=(parent-1)//2
                #print("  H  ", self.H)
                #print(" D ", self.D)
            print("HEAPIFY UP ENDS")    
            
            
        #heapify down
        elif child_2<self.size and self.D[self.H[ind]]<min(self.D[self.H[child_1]], self.D[self.H[child_2]]):
            print("HEAPIFY DOWN BEGINS")
            while child_2<self.size and self.D[self.H[ind]]<min(self.D[self.H[child_1]],self.D[self.H[child_2]]):
                max_ind=ind
                if self.D[self.H[child_1]]>self.D[self.H[ind]]:
                    max_ind=child_1
                elif self.D[self.H[child_2]]>self.D[self.H[ind]]:
                    max_ind=child_2
                if max_ind!=ind:
                    #D[H[max_ind]], D[H[ind]] = D[H[ind]], D[H[max_ind]]
                    self.node_heap[self.H[max_ind]], self.node_heap[self.H[ind]] = self.node_heap[self.H[ind]], self.node_heap[self.H[max_ind]]
                    self.H[max_ind], self.H[ind] = self.H[ind],self.H[max_ind]
                    ind=max_ind
                    child_1, child_2 = (2*ind)+1, (2*ind)+2

                elif max_ind==ind:
                    break
    
                #print("  H  ", self.H)
                #print(" D ", self.D)  
                
            if child_1<self.size and self.D[self.H[ind]]<self.D[self.H[child_1]]:
                #D[H[ind]], D[H[child_1]] = D[H[child_1]], D[H[ind]]
                self.node_heap[self.H[ind]], self.node_heap[self.H[child_1]] = self.node_heap[self.H[child_1]], self.node_heap[self.H[ind]]
                self.H[ind], self.H[child_1] = self.H[child_1], self.H[ind]

            print("HEAPIFY DOWN ENDS")    

        #iheapify down, if vertex has only one child,    
        elif child_1<self.size and self.D[self.H[ind]]<self.D[self.H[child_1]]:
            print("HEAPIFY DOWN WITH A SINGLE CHILD BEGINGS")
            self.node_heap[self.H[ind]], self.node_heap[self.H[child_1]] = self.node_heap[self.H[child_1]], self.node_heap[self.H[ind]]
            self.H[ind], self.H[child_1] = self.H[child_1], self.H[ind]
            print("HEAPIFY DOWN WITH A SINGLE CHILD ENDS")
            

        return self.H, self.D




    def max_heap(self):
        if self.size>0:
            return self.H[0], self.D[self.H[0]]
        return None

                
        
```
