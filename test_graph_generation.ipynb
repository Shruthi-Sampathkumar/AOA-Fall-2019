```python
import random
class Node(object):
    def __init__(self,v,w):
        self.vertex=v
        self.weight=w
```

```python
def graph_generation():
    '''
    number of vertices is 5000 numbered from 1 to 4999
    edge weights vary from 1 to 1000
    adjacency list for every vertex is list of 'Node' objects (contains vertex name and weight of the edge)
    '''
    
    degree = [int(random.normalvariate(6,1)) for i in range(5000)]
    
    #can be deleted
    #to check the average of the degree of edges 
    sum=0
    for i in range(5000):
        sum+=degree[i]
    print("Average degree of the edges is : ", sum/5000)
    print("\n")
    
    #creating a cycle, connecting the whole graph
    adj_list = [[] for _ in range(5000)]
    for i in range(1,5000):
        edge_weight = random.randint(1,1000)
        adj_list[i].append(Node(i-1,edge_weight))
        degree[i]-=1
        adj_list[i-1].append(Node(i,edge_weight))
        degree[i-1]-=1
        
    edge_weight = random.randint(1,1000)    
    adj_list[0].append(Node(4999,edge_weight))
    degree[0]-=1
    adj_list[4999].append(Node(0,edge_weight))
    degree[4999]-=1
    
    print("Graph is connected with a cycle")
    print("\n")
    
    #creating edges randomly until degree[vertex] becomes 0
    for vertex in range(5000):
        print("Currently checking vertex ", vertex)
        existence={}
        num_try=0
        for neigh in adj_list[vertex]:
            existence[neigh.vertex]=1
            
        while degree[vertex]>0:
            #print('num_try: ',num_try)
            new_neighbor = random.randint(0,4999)
            if degree[new_neighbor]>0 and new_neighbor not in existence:
                edge_weight=random.randint(1,1000)
                adj_list[vertex].append(Node(new_neighbor,edge_weight))
                degree[vertex]-=1
                adj_list[new_neighbor].append(Node(vertex,edge_weight))
                degree[new_neighbor]-=1

                
            elif num_try>=5 and new_neighbor not in existence:
                edge_weight=random.randint(1,1000)
                adj_list[vertex].append(Node(new_neighbor,edge_weight))
                degree[vertex]-=1
                adj_list[new_neighbor].append(Node(vertex,edge_weight))
                degree[new_neighbor]-=1
                
                num_try=0
                
                
            else:
                num_try+=1    
                
            existence[new_neighbor]=1 
    print("Graph Creation is completed")
    print("\n")
    return adj_list
                
    
```

```python
adjacency_list = graph_generation()
```
